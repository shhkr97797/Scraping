ニュース記事の取得

=============================================

```python
import requests
from bs4 import BeautifulSoup
import re
import time
```

```python
URL = "https://www.yahoo.co.jp/"
res = requests.get(URL)
```

```python
soup = BeautifulSoup(res.text, "html.parser")
```

```python
elms = soup.find_all(href=re.compile("https://news.yahoo.co.jp/pickup/"))
```

```python
pick_up = [i.attrs["href"] for i in elms] #リスト内包表記 【参考】https://qiita.com/shiho97797/items/4147f09b66d671cf530c
```

```python
for links in pick_up:
    res = requests.get(links)
    res_soup = BeautifulSoup(res.text, "html.parser")
    read_more_btn = res_soup.find("p", class_="sc-beKmYL")
    read_more_link = read_more_btn.a.attrs["href"]
    
    news_detail = requests.get(read_more_link)
    news_soup = BeautifulSoup(news_detail.text, "html.parser")
    print(news_soup.title.text)
    print(read_more_link + "\n====================================================\n")
    
    news_content = news_soup.find(class_=re.compile("sc-dliRfk"))
    print(news_content.text if hasattr(news_content, "text") else "", end="\n\n\n\n\n")
    
    time.sleep(2)
```
